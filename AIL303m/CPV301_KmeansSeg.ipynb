```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage import io
```

```python
# Load image from URL
image_url = "https://img.freepik.com/premium-photo/paradise-island-cinematic-photo-sunny-day-high-quality_599292-4013.jpg"
image = io.imread(image_url)

# Reshape the image to be a list of pixels
pixels = image.reshape((-1, 3))

# Convert to float for k-means input
pixels = np.float32(pixels)

# Define the number of clusters (segments)
n_clusters = 60

# Apply k-means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
kmeans.fit(pixels)

# Get the cluster labels for each pixel
labels = kmeans.labels_

# Get the center colors for each cluster
centers = np.uint8(kmeans.cluster_centers_)

# Create the segmented image
segmented_image = centers[labels].reshape(image.shape)

# Display the original and segmented images
plt.figure(figsize=(20, 9))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(segmented_image)
plt.title(f"Segmented Image ({n_clusters} clusters)")
plt.axis('off')

plt.show()
```
